# **Implementing TF-IDF from scratch**



```python
# Try Chaging the corpus and test the code
corpus = [
     'this is the first document mostly',
     'this document is the second document',
     'and this is the third one',
     'is this the first document here',
]
```

```python
import numpy as np
def computeTFIDF (corpus):
  # converting th corpus to lower case and finding the unique words from corpus and storing them as dimensionset
  dimensionset= sorted(set([word.lower() for word in set(str(" ".join(corpus)).split())]))
  tfidfvector=list()
  idfdict=dict()
  ''' calulating the idf for dimensionset'''
  for dimensionword in dimensionset:
      sentencecount=0
      for sentence in [sentence.split() for sentence in corpus]:
          if dimensionword in sentence:
              sentencecount+=1
      idfdict[dimensionword]=np.log(len(corpus)/sentencecount)
# for every sentence in corpus we are finding the Term Frequency Multiplying with the prior calculated IDF to get the TF-IDF 
  for i in range(len(corpus)):
      tfidfvector.append([])
      for word in (corpus[i].split()):
          termfrequency = (list(corpus[i].split()).count(word))/len(corpus[i].split())
          tfidf = idfdict[word] * termfrequency
          tfidfvector[i].append(tfidf.round(2))
  return np.array(tfidfvector)
```
