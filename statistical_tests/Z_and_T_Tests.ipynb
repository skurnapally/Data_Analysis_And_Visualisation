# Z-Test

Z-test is a statistical method to determine whether the distribution of the test statistics can be approximated by a normal distribution. It is the method to determine whether two sample means are approximately the same or different when their variance is known and the sample size is large (should be >= 30).

## When to Use Z-test:

1. The sample size should be greater than 30. Otherwise, we should use the t-test.
2. Samples should be drawn at random from the population.
3. The standard deviation of the population should be known.
4. Samples that are drawn from the population should be independent of each other.
5. The data should be normally distributed, however for large sample size, it is assumed to have a normal distribution.

## Assumptions of z-test:

1.     Means of the two sample random variables which are compared follow normal distribution.
2.     Standard deviation of the population (from which R.V are drawn) of the random variables which are compared are known or number of samples in those R.V are reasonably large that means at least above 30. So that we can assume sample standard deviations as the population standard deviations. 

## Task: 
* comparing the means of the two R.V

## Null and alternative hypothesis:
* H0 – sample means are equal ( ideally population means are equal)
* Ha – sample means are not equal

## Type of test :

* 2 sided test.(Not always it depends on the alternative hypothesis)

## P-Value:

* Area under the N(0,1) distribution curve on both sides of the Tobs as it is two sided test.
* And if p-value is less then significance level we reject null hypothesis at that significance level.







### One Sample Z-Test

* We can use the ztest() function from the statsmodels package to perform one sample and two sample z-tests in Python.

Reference: https://www.statsmodels.org/dev/generated/statsmodels.stats.weightstats.ztest.html

<font color = 'red'><b>statsmodels.stats.weightstats.ztest(x1, x2=None, value=0)</b></font>

where:

x1: values for the first sample

x2: values for the second sample (if performing a two sample z-test)

value: mean under the null (in one sample case) or mean difference (in two sample case)

alternative: The alternative hypothesis, H1, has to be one of the following

* ‘two-sided’: H1: difference in means not equal to value (default) 
* ‘larger’ : H1: difference in means larger than value 
* ‘smaller’ : H1: difference in means smaller than value

one sample test with example:

> Suppose the IQ in a certain population is normally distributed with a mean of μ = 100 and standard deviation of σ = 15.

> A researcher wants to know if a new drug affects IQ levels, so he recruits 1000 patients to try it and records their IQ levels.

> The following code shows how to perform a one sample z-test in Python to determine if the new drug causes a significant difference in IQ levels:

```python
from statsmodels.stats.weightstats import ztest as ztest
import numpy as np

#enter IQ levels for 50 patients
data = np.random.normal(100, 15, 1000)

#perform one sample z-test
ztest_Score, p_value = ztest(data, value=100,alternative='two-sided')
print(ztest_Score)
print(p_value)
```

As P-Value is greater than 0.05 significance level we fail to reject the null hypothesis.

# Two sample Z-Test

> Suppose the IQ levels among individuals in two different cities are known to be normally distributed with known standard deviations.

> A researcher wants to know if the mean IQ level between individuals in city A and city B are different, so he/she selects a simple random sample of  1000 individuals from each city and records their IQ levels.

> The following code shows how to perform a two sample z-test in Python to determine if the mean IQ level is different between the two cities:

```python
from statsmodels.stats.weightstats import ztest as ztest

#enter IQ levels for 20 individuals from each city
cityA = np.random.normal(100, 15, 1000)

cityB = np.random.normal(120, 30, 1000)

#perform two sample z-test
ztest_Score, p_value = ztest(cityA, cityB, value=0,alternative = 'two-sided')
# 'two-sided' - IQ of city 

print(ztest_Score)
print(p_value)
```

## Z_Proportions-Test

Reference: https://www.statsmodels.org/dev/generated/statsmodels.stats.proportion.proportions_ztest.html

* Z-test is used to compare means and z-proportions test is used for comparing the proportions or ratios.

### Task: 
* comparing the proportions of the two R.V

### Null and alternative hypothesis:
* H0 –proportions are equal
* Ha – proportions are not equal

### Assumptions:
* We can use this test only when we know the distribution of the Test statistic.

### Test Statistic:

image.png

* And test statistic follows Z(0,1) hence z-proportions test name.

### Type of test :

* 2 sided test.(Not always it depends on the alternate hypothesis)

### P-Value:
* Area under the N(0,1) curve on both sides of the Tobs as it is two sided test



### One sampled Z_proportionstest

```python
from statsmodels.stats.proportion import proportions_ztest

# can we assume anything from our sample
significance = 0.05

# our sample - 82% are good
sample_success = 410
sample_size = 500

# our Ho is  80%
null_hypothesis = 0.80

# check our sample against Ho for Ha > Ho
# for Ha < Ho use alternative='smaller'
# for Ha != Ho use alternative='two-sided'
stat, p_value = proportions_ztest(count=sample_success, nobs=sample_size, value=null_hypothesis, alternative='larger')# right sided test

# report
print('z_stat: %0.3f, p_value: %0.3f' % (stat, p_value))

if p_value > significance:
   print ("Fail to reject the null hypothesis ")
else:
   print ("Reject the null hypothesis - suggest the alternative hypothesis is true")
```

## Two sampled z_proportions_test

```python
from statsmodels.stats.proportion import proportions_ztest
import numpy as np

# can we assume anything from our sample
significance = 0.025

# our samples - 82% are good in one, and ~79% are good in the other
# note - the samples do not need to be the same size
sample_success_a, sample_size_a = (410, 500)
sample_success_b, sample_size_b = (379, 400)

# check our sample against Ho for Ha != Ho
successes = np.array([sample_success_a, sample_success_b])
samples = np.array([sample_size_a, sample_size_b])

# note, no need for a Ho value here - it's derived from the other parameters
stat, p_value = proportions_ztest(count=successes, nobs=samples,  alternative='two-sided')

# report
print('z_stat: %0.3f, p_value: %0.3f' % (stat, p_value))

if p_value > significance:
   print ("Fail to reject the null hypothesis")
else:
   print ("Reject the null hypothesis - suggest the alternative hypothesis is true")
```

# T-Test

## Assumptions of T-test:
1.     We can use this test only when we know the distribution of the Test statistic.
2.     Means of the two sample random variables which are compared follow normal distribution.
3.     Standard deviation of the population (from which R.V are drawn) of the random variables which are compared are unknown or number of samples in those R.V are not reasonably large that means below 30. 
4.     Samples are random and independent of one another.
## Task: 
* comparing the means of the two R.V

## Null and alternative hypothesis:
* H0 – sample means are equal ( ideally population means are equal)
* Ha – sample means are not equal

## Test Statistic:

image.png
￼
* Above is the test statistic we use sample means and sample standard deviations.
*  And test statistic will follow t-distribution with some degrees of freedom(v) , and those degrees of freedom in t-test is n+m-2 , and when the number of samples increases then degrees of freedom increases when degrees of freedom increases as we can see from the t-distribution pdf , t-distribution will become z distribution (N(0,1)) and hence t-test will become z-test.

## Type of test :
* 2 sided test.(Not always it depends on the alternative hypothesis)

## P-Value:
* Area under the t-distribution curve on both sides of the Tobs as it is two sided test.
* And if p-value is less then significance level we reject null hypothesis at that significance level.

## Two Sample T-Test

Reference: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html#scipy.stats.ttest_ind

```python
from scipy import stats as st
from bioinfokit.analys import get_data
# load dataset as pandas dataframe
df = get_data('t_ind_samp').data
df.head(2)
```
